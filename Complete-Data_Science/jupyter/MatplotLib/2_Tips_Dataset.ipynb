```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r'E:\DataScience\MatplotLib\Bank_churn.csv')
```

```python
data.head()
```

## 1(i). Scatter-Plot

```python
data.plot.scatter(x= 'Age',y='Balance')
# Insights in data
# 1 Maximum concenration >> 30 - 50
# 2 Maximum Salary >> 40 age
# 3 Minimum Salary >> all 
```

## 1(ii). Histplot

### Age and Balance Distribution

```python
# Age Distribution
plt.title("Age distribution")
plt.hist(data["Age"])
plt.show()

# Balance Distribution

plt.title("Balance Distribution")
plt.hist(data["Balance"],color='red')
plt.show()
```

## 2. Violinplot

```python
plt.violinplot(data['Balance'])
plt.show()
```

## 3 Pie Plot

```python
langs = ['C++','Java','Python','C']
students = [10,40,30,20]
```

### (a). Making simple pie plot

```python
plt.pie(students,labels = langs)
plt.show()
```

### (b). Adding percentage and shadow

```python
plt.pie(students,labels= langs,autopct = '%1.1f%%',shadow = True)
plt.show()
```

### (c). Highlighting a Slice or exploding

```python
my_explode = (0.1,0.0,0.0,0.0)
plt.pie(students,labels = langs,autopct = '%1.1f%%',shadow=True,explode = my_explode)
plt.show()
```
