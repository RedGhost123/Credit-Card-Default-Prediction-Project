### 2(vii).
## (a)
####  Adding Grid in plot and tittle

```python
plt.plot(x,np.sin(x-1),color='red')
plt.ylim(-1.5,1.5)
plt.xlim(1.15)
plt.axis('equal') # Making equal axis for filling full page
plt.grid() # Adding grid
plt.title("this is sin plot before execution of real dataset")  # Adding title
plt.show()
```

## (b)
#### Adding Label of X and Y axis and adding sin and cos x with lable

```python
plt.plot(x,np.sin(x-1),color = 'red',label = 'Sin-X')
plt.plot(x,np.cos(x-1),color = 'blue',label = 'Cos-X')
plt.xlim(1,15)
plt.ylim(-1.5,1.5)
#plt.axis('equal')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid()
plt.legend()  # Legen function gives the information about label inside plot
```

## 3(i). Dividing a page in 2 part along Columns

#### plt.subplot(1,2,1) >> [1] 1 is Row, [2] 2 is Columns and [3] 1 is ploting figure in 1st column

```python
plt.figure()  # Ploting page
plt.subplot(1,2,1) # 1- page, 2 - making 2 sub-page, ploting graph in 1 subpage
plt.plot(x,np.sin(x))
```

```python
plt.figure()
plt.subplot(1,2,1)
plt.plot(x,np.sin(x))

plt.subplot(1,2,2) # making plot in 2nd page
plt.plot(x,np.cos(x)) # ploting graph
```

## 3(ii). Dividing a page in 2 part along Rows

```python
plt.figure()
plt.subplot(2,1,1)
plt.plot(x,np.sin(x))

plt.subplot(2,1,2)
plt.plot(x,np.cos(x))
```

## 3(iii). Dividing a page in 4 Parts and also Changing Colours

```python
plt.figure()
plt.subplot(2,2,1)
plt.plot(x,np.sin(x), color = 'blue')

plt.subplot(2,2,2)
plt.plot(x,np.cos(x), color = 'green')

plt.subplot(2,2,3)
plt.plot(x,np.tan(x), color = 'red')

plt.subplot(2,2,4)
plt.plot(x,np.sin(x-1), color = 'skyblue')
```

### 5(i). Ploting Scatter Plot

```python
x= np.random.rand(50)
y= np.random.rand(50)
plt.figure(figsize=(6,4)) # x = 6 and y = 4
plt.scatter(x,y,c='r',alpha=0.8,marker="*") # we can remove marker it is only used to highlight, Alpha is intensity of color
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot")
plt.show()
```

### 7. Ploting 3-D plot

```python
x1 = np.random.rand(20)
y1 = np.random.rand(20)
z1 = np.random.rand(20)
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.scatter(x1,y1,z1,color = 'b', alpha = 0.8)
plt.ylabel("Y-axis")
plt.xlabel("X-axis")
plt.show()
```
