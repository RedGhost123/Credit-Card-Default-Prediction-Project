```python
def kde(col):
    grid=sns.FacetGrid(df_copy, hue = "class", height=6, aspect=2)  # FacetGrid >> Page info
    grid.map(sns.kdeplot,col)
    grid.add_legend()
```

```python
kde('red_blood_cell_count')
```

```python
kde('white_blood_cell_count')
```

```python
kde('haemoglobin')
```

```python
def violin(col):
    fig = px.violin(df_copy,y = col,x='class',box = True,template = 'plotly_dark')
    return fig.show()
```
