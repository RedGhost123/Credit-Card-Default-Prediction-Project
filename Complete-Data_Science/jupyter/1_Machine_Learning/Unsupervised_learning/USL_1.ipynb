### K-Means Clustering


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_blobs
```

```python
x,y = make_blobs(n_samples=1000,centers=3,n_features=2,random_state=1)
```

```python
x
```

```python
y
```

```python
plt.scatter(x[:, 0],x[:,1], c = y)
```

### Train_Test split

```python
from sklearn.model_selection import train_test_split
X_train,X_Test,y_train,y_test = train_test_split(x,y,test_size = 0.30,random_state = 1)
```

```python
X_train.shape,X_Test.shape
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 3)
```

```python
kmeans
```

```python
kmeans.fit(X_train)
```

```python
y_labels = kmeans.predict(X_train)
```

```python
plt.scatter(X_train[:,0],X_train[:,1], c= y_labels)
```

```python
from sklearn.metrics import silhouette_score
silhouette_score(X_train, kmeans.labels_)
kmeans.labels_
```

```python
kmeans.inertia_
```

### To select K(centeroid)  > elbow method

```python
wcss = []
for k in range(1,10):
    kmeans  = KMeans(n_clusters=k)
    kmeans.fit(X_train)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

### Plot an elbow curve

```python
plt.plot(range(1,10),wcss)
plt.xticks(range(1,10))
plt.xlabel("no of clusters")
plt.ylabel("wcss")
plt.show()
# Based on the above plot the optimal number of k cluser is 3
```

using kneed for elbow locator

```python
from kneed import KneeLocator
k1= KneeLocator(range(1,10),wcss,curve='convex',direction='decreasing')
```

```python
k1.elbow
```

### Silhoute coeff

```python
silhouette_coeff = []
for k in range(2,10):
    kmeans  = KMeans(n_clusters=k)
    kmeans.fit(X_train)
    score = silhouette_score(X_train,kmeans.labels_)
    silhouette_coeff.append(score)
```
