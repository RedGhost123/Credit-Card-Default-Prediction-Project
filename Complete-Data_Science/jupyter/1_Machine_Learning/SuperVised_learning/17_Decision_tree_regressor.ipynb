```python
import pandas as pd
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
import warnings 
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import fetch_california_housing
```

```python
data = fetch_california_housing()
```

```python
data.keys()
```

```python
df = pd.DataFrame(data.data,columns=data.feature_names)
```

```python
df
```

```python
df['price'] = data.target
```

```python
df['price']
```

```python
df
```

```python
x = df.iloc[:,:-1]
```

```python
x
```

```python
y = df.iloc[:,-1]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state=1)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dtr_model = DecisionTreeRegressor()
```

```python
dtr_model
```

```python
dtr_model.fit(x_train,y_train)
```

```python
y_pred = dtr_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,r2_score
```

```python
r2_score(y_test,y_pred)
```

### Hyperparameter Tuning

```python
parameter = {
    'criterion':['squared_error','poissor'],
    'max_depth':[1,2,3,4,6,8],
    'splitter':['best','random'],
    'max_features':['sqrt','log2']
}
```

```python
parameter
```

```python
reg = DecisionTreeRegressor()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
model = GridSearchCV(reg,param_grid=parameter,cv = 5,scoring="neg_mean_squared_error")
```

```python
model
```

```python
model.fit(x_train,y_train)
```

```python
model.best_params_
```

```python
final_model = DecisionTreeRegressor(criterion='squared_error',max_depth=8)
```

```python
final_model
```

```python
final_model.fit(x_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(12,12))
tree.plot_tree(final_model,filled=True)
plt.show()
```
