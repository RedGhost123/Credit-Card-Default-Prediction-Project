### 1.(i) Merge and Joint Function

```python
df_3 = pd.DataFrame({'key1':[1,2,3,4,5],
                  'key2':[4,5,6,7,8],
                  'key3':[5,6,7,8,9]})
```

```python
df_3
```

```python
df_4 = pd.DataFrame({'key1':[2,3,6,8,9],
                     'key4':[42,3,2,68,28],
                     'key5':[78,8,9,34,11]})
```

```python
df_4
```

### 1(ii). Merge Function (Inner Joint)

```python
pd.merge(df_3,df_4,how='left')
```

```python
pd.merge(df_3,df_4,how='right')
```

### 1(iii). Merge Function (outer joint)

```python
pd.merge(df_3,df_4,how='outer')
```

### 1(iv). Merge Function (cross Joint)

```python
pd.merge(df_3,df_4,how='cross')
```

### 1(v) Merging function with common columns and data

```python
pd.merge(df_3,df_4,how='left',left_on='key2',right_on='key4')
```

## 2. Joint Function

```python
df_5 = pd.DataFrame({'key1':[1,2,45,6,67],
                     'key2':[56,5,6,7,8],
                     'key3':[3,56,5,6,6]},index = ['a','b','c','d','e'])
```

```python
df_6= pd.DataFrame({'key6':[1,2,3,4,5],
                    'key4':[4,5,6,7,8],
                    'key5':[3,4,5,6,6]},index = ['a','b','h','i','j'])
```

```python
df_5
```

```python
df_6
```

### 2(i) Joining DF_5 and DF_6   [Make sure Columns name should not be Same. ]

```python
df_5.join(df_6,how='left') # Left Join
```

```python
df_5.join(df_6,how='right') # Right Join
```

```python
df_5.join(df_6,how='inner') # Inner Join
```

```python
df_5.join(df_6,how='outer') # Outer Join
```

```python
df_5.join(df_6,how='cross') # Cross Join
```
