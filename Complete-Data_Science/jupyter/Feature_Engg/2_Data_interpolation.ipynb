```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Data Interpolation  >> Is a process of estimating data.
# 1. Linear Interpolation 
# 2. Cubic Interpolation
# 3. Polynomial Interpolation
x = np.array([1,2,3,4,5])
y = np.array([1,3,5,7,8])
```

#### Data Interpolation is used to create a senthatic data point or artificial data.

```python
plt.scatter(x,y)
```

### 1. Linear Interpolate

```python
x_new = np.linspace(1,5,10)
```

```python
x_new
```

```python
y_interp = np.interp(x_new,x,y)
```

```python
y_interp
```

```python
plt.scatter(x_new,y_interp)
```

### 2. Cubic interpolation

```python
x = np.array([1,2,3,4,5])
y = np.array([1,8,27,64,125])
```

```python
plt.scatter(x,y)
```

#### Installing scipy > pip install scipy

```python
# Installing scipy >> pip install scipy

# Improrting scipy library
from scipy.interpolate import interp1d
```

```python
f  = interp1d(x,y,kind='cubic')
```

```python
f
```

```python
x1_new = np.linspace(1,5,10)
```

```python
y1_interp = f(x1_new)
```

```python
y1_interp
```

```python
plt.scatter(x1_new,y1_interp)
```

### 3. Polynomial Interpolation

```python
x2 = np.array([1,2,3,4,5])
y2 = np.array([1,4,8,1,25])
```

```python
plt.scatter(x2,y2)
```

```python
# polyfit > Polynomial interpolation
p = np.polyfit(x2,y2,2)
```

```python
p
```

```python
x2_new = np.linspace(1,5,10)
y2_interp = np.polyval(p,x2_new) # polyval > it will give you value
```

```python
x2_new
```

```python
y2_interp
```

```python
plt.scatter(x2_new,y2_interp)
```
