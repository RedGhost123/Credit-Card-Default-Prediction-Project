```python
import numpy as np 
import pandas as pd
```

### Creating Data

```python
np.random.seed(1) # For reproducibility
```

```python
no_samples = 1000
class_0_ratio = 0.9
no_class_0 =int(no_samples * class_0_ratio)
no_class_1 = 100
```

```python
class_0_ratio
```

```python
no_class_0
```

```python
no_class_1
```

```python
generated_number_c0 = np.random.normal(0,1, no_class_0)
generated_number_c1 = np.random.normal(3,1,no_class_1)
```

```python
generated_number_c0
```

```python
generated_number_c1
```

### Making Data frame of class 0

```python
[0]* 5
```

```python
class_0 = {'feature1': generated_number_c0,
 'feature2':generated_number_c0,
 'target':[0]*no_class_0}
```

```python
class_0 =  pd.DataFrame(class_0)
```

```python
class_0
```

```python
class1 = pd.DataFrame({'feature1':generated_number_c1,
                       'feature2':generated_number_c1,
                       'target':[1]*no_class_1})
```

```python
class1
```

```python
df = pd.concat([class_0,class1]).reset_index(drop = True)
```

```python
df
```

```python
df.target.value_counts()#(normalize=True)
```

```python
df.target.value_counts(normalize=True)
```

```python
df_minority = df[df.target == 1]
df_majority = df[df.target == 0]
```

```python
df_majority
```

```python
df_minority
```

### Installing Scikit-learn

```python
# UpSampling >> Increasing the minority data to majority number
from sklearn.utils import resample
```

### Resampling | UpSampling

```python
df_minority_upsampled = resample(df_minority, replace=True,  n_samples=len(df_majority), random_state=1)
```

```python
df_minority_upsampled
```

```python
df_minority_upsampled.duplicated().sum()
```

```python
df_upsampled = pd.concat([df_majority,df_minority_upsampled])
```

```python
df_upsampled
```

### DownSampling

```python
df_majority_downsampled = resample(df_majority, replace=False, n_samples=len(df_minority),random_state=1)
```

```python
df_majority_downsampled
```

```python
df_downsampled = pd.concat([df_minority,df_majority_downsampled])
```

```python
df_downsampled
```

### SMoTE (Synthetic Minority oversampling Technique)

```python
# SMoTE
```

```python
from sklearn.datasets import make_classification
```

#### Generating dataSet

```python
x,y = make_classification(n_samples=1000, n_redundant= 0, n_features=2, n_clusters_per_class= 1, weights=[0.90],random_state=1)
```

```python
x
```

```python
y
```

```python
len(y[y == 0])
```

```python
len(y[y==1])
```

```python
df1 = pd.DataFrame(x,columns=['f1','f2'])
df2 = pd.DataFrame(y,columns=['target'])
df_final = pd.concat([df1,df2],axis=1)
```

```python
df_final
```

```python
df_final.target.value_counts()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df_final['f1'], df_final['f2'], c = df_final['target'])
```

### Performing SMote

#### oversampling

```python
from imblearn.over_sampling import SMOTE
```

```python
oversample = SMOTE()
```

```python
x1,y1 = oversample.fit_resample(df_final[['f1','f2']], df_final['target'])
```

```python
x1
```

```python
y1
```

```python
len(y1[y1==0])
```

```python
len(y1[y1 == 1])
```

#### Making DataFrame

```python
df1 = pd.DataFrame(x1,columns=['f1','f2'])
df2 = pd.DataFrame(y1,columns=['target'])
oversampling_data = pd.concat([df1,df2],axis=1)
```

```python
plt.scatter(oversampling_data['f1'], oversampling_data['f2'], c = oversampling_data['target'])
```

#### Checking difference in Both data

```python
plt.figure(figsize = (12,6))
plt.subplot(1,2,1)
plt.scatter(df_final['f1'], df_final['f2'], c = df_final['target'])
plt.title("Before Smote")

plt.subplot(1,2,2)
plt.scatter(oversampling_data['f1'], oversampling_data['f2'], c = oversampling_data['target'])
plt.title("After Smote")

plt.show()
```
