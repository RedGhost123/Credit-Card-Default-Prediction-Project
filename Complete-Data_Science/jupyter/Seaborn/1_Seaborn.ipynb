## 3. Pie chart

```python
# Making simple pie plot

langs = ['python','java','C++','c']
students = [20,100,40,60]
plt.pie(students,labels=langs)
plt.show()
```

### (b) Adding percentage

```python
plt.pie(students,labels = langs,autopct = '%1.1f%%')
plt.show()
```

### Enhancing with shadow and highlighting with explode function

```python
my_explode = (0.1,0.0,0,0.0) # It is used to just slicing and highlighting
plt.pie(students,autopct='%1.1f%%',labels=langs,explode=my_explode,shadow=True) # Adding languages lable -> labels = langs, adding percentage -> %1.1f%%
plt.title("Pie Chart") # Adding title
plt.show()
```
